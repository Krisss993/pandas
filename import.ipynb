```python
url = 'https://ec.europa.eu/eurostat/estat-navtree-portlet-prod/BulkDownloadListing?file=data/avia_par_pl.tsv.gz&unzip=true'

# Download the gzip file
response = requests.get(url)
compressed_data = BytesIO(response.content)

# Decompress the gzip file and read it with Pandas
with gzip.open(compressed_data, 'rt', encoding='utf-8') as f:
    data = pd.read_csv(f, delimiter='\t')
    
```

```python
data.head(10)
```

```python
data_desc = data['unit,tra_meas,airp_pr\\time'].str.split(',', expand=True)
```

```python
data_desc.rename(columns={0:'unit',1:'tra_meas',2:'airpopr'}, inplace=True)
```

```python
data_desc.head()
```

```python
data = data_desc.join(data)
```

```python
data.head()
```

```python
data.drop('unit,tra_meas,airp_pr\\time', axis=1, inplace=True)
```
